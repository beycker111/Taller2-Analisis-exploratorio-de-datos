```python
#Limpiamos que sea ESTU_ESTADO publicar
#Limpiamos para que sean solo estudiantes colombianos

class ProcesamientoArchivos:
    
    @staticmethod
    def listosParaPublicarse(df):
        return df[(df['ESTU_ESTADO']=='PUBLICAR')]
    
    @staticmethod
    def nacionalidadColombiana(df):
        return df[(df['ESTU_NACIONALIDAD']=='1 COLOMBIA')]
    

df_filtrado = df_filtrado.pipe(ProcesamientoArchivos.listosParaPublicarse).pipe(ProcesamientoArchivos.nacionalidadColombiana)
df_filtrado['ESTU_NACIONALIDAD'].value_counts()
```

```python
df_filtrado['ESTU_FECHANACIMIENTO'] = pd.to_datetime(df_filtrado['ESTU_FECHANACIMIENTO'])

df_filtrado['ESTU_FECHANACIMIENTO_YEAR'] = df_filtrado["ESTU_FECHANACIMIENTO"].apply(lambda x: x.year)
df_filtrado['ESTU_FECHANACIMIENTO_YEAR']
```

```python
df_filtrado["ESTU_ETNIA"] = df_filtrado["ESTU_ETNIA"].fillna("100 Ninguno")
```

```python
#Removemos los acentos de ESTU_ETNIA
import unidecode

def remove_accents(a):
    return unidecode.unidecode(a)

df_filtrado['ESTU_ETNIA'] = df_filtrado['ESTU_ETNIA'].apply(remove_accents)
```

```python
tabla_freq_etnia=(pd.crosstab(index=df_filtrado["ESTU_ETNIA"],columns="count")).reset_index()
tabla_freq_etnia['Freq. Rel.']=tabla_freq_etnia['count']/sum(tabla_freq_etnia['count'])
tabla_freq_etnia.rename(columns={'count':'Freq. Abs.'},inplace=True)
tabla_freq_etnia=tabla_freq_etnia.sort_values(by='Freq. Abs.',ascending=False).reset_index(drop=True)
tabla_freq_etnia
```

```python
def transformarLimitaciones(limitacion):
    limitacion = limitacion.fillna("NO")
    limitacion = limitacion.replace("x", "SI")
    return limitacion

df_filtrado["ESTU_LIMITA_MOTRIZ"] = transformarLimitaciones(df_filtrado["ESTU_LIMITA_MOTRIZ"])
df_filtrado["ESTU_LIMITA_INVIDENTE"] = transformarLimitaciones(df_filtrado["ESTU_LIMITA_INVIDENTE"])
df_filtrado["ESTU_LIMITA_CONDICIONESPECIAL"] = transformarLimitaciones(df_filtrado["ESTU_LIMITA_CONDICIONESPECIAL"])
df_filtrado["ESTU_LIMITA_SORDO"] = transformarLimitaciones(df_filtrado["ESTU_LIMITA_SORDO"])
df_filtrado["ESTU_LIMITA_AUTISMO"] = transformarLimitaciones(df_filtrado["ESTU_LIMITA_AUTISMO"])

df_limitaciones = df_filtrado[["ESTU_LIMITA_MOTRIZ", "ESTU_LIMITA_INVIDENTE", "ESTU_LIMITA_CONDICIONESPECIAL", "ESTU_LIMITA_SORDO","ESTU_LIMITA_AUTISMO"]]
df_limitaciones
```

```python
#count_rows = (df_filtrado["ESTU_LIMITA_MOTRIZ"] == "SI").sum()
limitaciones = ("MOTRIZ", "INVIDENTE", "CONDICIONESPECIAL", "SORDO", "AUTISMO")
limitacion_categorizadas = {
    'LIMITACIÓN': limitaciones,
    'SI (Fecuencia Relativa %)': ((df_filtrado["ESTU_LIMITA_MOTRIZ"] == "SI").sum(),
           (df_filtrado["ESTU_LIMITA_INVIDENTE"] == "SI").sum(),
          (df_filtrado["ESTU_LIMITA_CONDICIONESPECIAL"] == "SI").sum(),
          (df_filtrado["ESTU_LIMITA_SORDO"] == "SI").sum(),
          (df_filtrado["ESTU_LIMITA_AUTISMO"] == "SI").sum()),
    'NO (Fecuencia Relativa %)': ((df_filtrado["ESTU_LIMITA_MOTRIZ"] == "NO").sum(),
          (df_filtrado["ESTU_LIMITA_INVIDENTE"] == "NO").sum(),
          (df_filtrado["ESTU_LIMITA_CONDICIONESPECIAL"] == "NO").sum(),
          (df_filtrado["ESTU_LIMITA_SORDO"] == "NO").sum(),
          (df_filtrado["ESTU_LIMITA_AUTISMO"] == "NO").sum())
}
limitacion_categorizadas = pd.DataFrame(limitacion_categorizadas)

limitacion_categorizadas["SI (Fecuencia Relativa %)"] = limitacion_categorizadas["SI (Fecuencia Relativa %)"].apply(lambda x: round((x/len(df_limitaciones))*100,2))
limitacion_categorizadas["NO (Fecuencia Relativa %)"] = limitacion_categorizadas["NO (Fecuencia Relativa %)"].apply(lambda x: round((x/len(df_limitaciones))*100,2))

limitacion_categorizadas
```

```python
limitaciones = ("MOTRIZ", "INVIDENTE", "CONDICIONESPECIAL", "SORDO", "AUTISMO")

#penguin_means = df_filtrado[["ESTU_LIMITA_MOTRIZ", "ESTU_LIMITA_INVIDENTE", "ESTU_LIMITA_CONDICIONESPECIAL", "ESTU_LIMITA_SORDO","ESTU_LIMITA_AUTISMO"]]

si_no = {
    'SI': tuple(list(limitacion_categorizadas["SI (Fecuencia Relativa %)"])),
    'NO': tuple(list(limitacion_categorizadas["NO (Fecuencia Relativa %)"]))
}

x = np.arange(len(limitaciones))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in si_no.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frecuencia Absoluta')
ax.set_title('Estudiantes con limitaciones')
ax.set_xticks(x + width, limitaciones)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 130)

plt.show()
```
